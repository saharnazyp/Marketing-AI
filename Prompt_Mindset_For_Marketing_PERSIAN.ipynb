## ۱. مسئله‌ای که حل می‌کنیم
تولید محتوا زیاد است، ولی **برند ندارد**.  
هدف: محتوای اثرگذار بسازیم که هم ارزش برند را منتقل کند و هم فروش ایجاد کند.

---

## ۲. پرامپت برای کپی فروش (Copywriting)

### ساختار محتوا
1. **تیتر (Headline)**: کوتاه، جذاب، مستقیم  
2. **بدنه (Body)**: توضیح مزیت یا محصول، نشان دادن نیاز مخاطب  
3. **CTA (Call To Action)**: مستقیم و واضح، مخاطب را به اقدام دعوت می‌کند  

### تمرین



```python
# محصول یا سرویس خود را وارد کنید
product_name = "مثال: پرامپت‌های فروش‌ساز"

# تیتر
headline = "تیم مارکتینگت محتوا تولید می‌کنه ولی فروش نمی‌سازه؟"

# بدنه
body = f"با {product_name}، محتواهایی بساز که نه فقط دیده بشن، بلکه فروش واقعی ایجاد کنند."

# CTA
cta = "همین الان دریافت کن"

# نمایش محتوا
print("تیتر:", headline)
print("بدنه:", body)
print("CTA:", cta)
```


**پرامپت برای Storytelling**

ساختار محتوا

مشکل (Problem): درد یا نیاز مخاطب

همدلی (Empathy): نشان دادن درک برند از مشکل

راه‌حل (Solution): معرفی محصول یا سرویس

```python
problem = "تلاش می‌کنید محتوا بسازید اما نتیجه نمی‌گیرید؟"
empathy = "ما می‌دانیم چقدر ناامیدکننده است وقتی تلاش می‌کنید اما فروش نمی‌بینید."
solution = f"با {product_name}، محتوای شما به فروش تبدیل می‌شود."

print("مشکل:", problem)
print("همدلی:", empathy)
print("راه‌حل:", solution)
```

**پرامپت برای ارزش پیشنهادی (Value Proposition)**

```python
value_proposition = "راهی سریع برای افزایش فروش بدون نیاز به صرف زمان زیاد."
print("ارزش پیشنهادی:", value_proposition)
```

**کنترل لحن (Tone of Voice)**

لحن‌های قابل انتخاب:

لوکس

دوستانه

جوان

رسمی سازمانی

```python
tone = "دوستانه"  # تغییر دهید: لوکس / دوستانه / جوان / رسمی سازمانی
print(f"لحن انتخاب شده: {tone}")
```

**تمرین عملی**

برای یک محصول مشخص، سه لحن، سه مخاطب، سه هدف بسازید:

```python
# محصولات
product = "پرامپت‌های فروش‌ساز"

# لحن‌ها
tones = ["لوکس", "دوستانه", "جوان"]

# مخاطب‌ها
audiences = ["جدید", "آگاه", "حرفه‌ای"]

# اهداف
goals = ["آگاهی", "فروش", "وفاداری"]

for t in tones:
    for a in audiences:
        for g in goals:
            print(f"\n---\nلحن: {t} | مخاطب: {a} | هدف: {g}")
            print(f"تیتر: {product} مناسب برای {a} با هدف {g}")
            print(f"CTA: همین الان امتحان کن!")
```
