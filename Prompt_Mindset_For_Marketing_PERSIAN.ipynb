## ۲. پرامپت برای کپی فروش (Copywriting)

### ساختار محتوا
1. **تیتر (Headline)**: کوتاه، جذاب، مستقیم  
2. **بدنه (Body)**: توضیح مزیت یا محصول، نشان دادن نیاز مخاطب  
3. **CTA (Call To Action)**: مستقیم و واضح، مخاطب را به اقدام دعوت می‌کند  

### تمرین



```python
# محصول یا سرویس خود را وارد کنید
product_name = "مثال: پرامپت‌های فروش‌ساز"

# تیتر
headline = "تیم مارکتینگت محتوا تولید می‌کنه ولی فروش نمی‌سازه؟"

# بدنه
body = f"با {product_name}، محتواهایی بساز که نه فقط دیده بشن، بلکه فروش واقعی ایجاد کنند."

# CTA
cta = "همین الان دریافت کن"

# نمایش محتوا
print("تیتر:", headline)
print("بدنه:", body)
print("CTA:", cta)
```


**پرامپت برای Storytelling**

ساختار محتوا

مشکل (Problem): درد یا نیاز مخاطب

همدلی (Empathy): نشان دادن درک برند از مشکل

راه‌حل (Solution): معرفی محصول یا سرویس

```python
problem = "تلاش می‌کنید محتوا بسازید اما نتیجه نمی‌گیرید؟"
empathy = "ما می‌دانیم چقدر ناامیدکننده است وقتی تلاش می‌کنید اما فروش نمی‌بینید."
solution = f"با {product_name}، محتوای شما به فروش تبدیل می‌شود."

print("مشکل:", problem)
print("همدلی:", empathy)
print("راه‌حل:", solution)
```

**پرامپت برای ارزش پیشنهادی (Value Proposition)**

```python
value_proposition = "راهی سریع برای افزایش فروش بدون نیاز به صرف زمان زیاد."
print("ارزش پیشنهادی:", value_proposition)
```

**کنترل لحن (Tone of Voice)**

لحن‌های قابل انتخاب:

لوکس

دوستانه

جوان

رسمی سازمانی

```python
tone = "دوستانه"  # تغییر دهید: لوکس / دوستانه / جوان / رسمی سازمانی
print(f"لحن انتخاب شده: {tone}")
```

**تمرین عملی**

برای یک محصول مشخص، سه لحن، سه مخاطب، سه هدف بسازید:

```python
# محصولات
product = "پرامپت‌های فروش‌ساز"

# لحن‌ها
tones = ["لوکس", "دوستانه", "جوان"]

# مخاطب‌ها
audiences = ["جدید", "آگاه", "حرفه‌ای"]

# اهداف
goals = ["آگاهی", "فروش", "وفاداری"]

for t in tones:
    for a in audiences:
        for g in goals:
            print(f"\n---\nلحن: {t} | مخاطب: {a} | هدف: {g}")
            print(f"تیتر: {product} مناسب برای {a} با هدف {g}")
            print(f"CTA: همین الان امتحان کن!")
```

CTA اثرگذار طراحی کند

Storytelling و ارزش پیشنهادی را در متن اعمال کند

# جلسه سوم

**پرامپت تصویر و ویدیو (Visual Prompting)**

چرا این فاز مهم است؟

در دنیای امروز، تصویر قبل از متن دیده می‌شود.

مغز انسان تصویر را چندین برابر سریع‌تر از متن پردازش می‌کند.

بیشتر برندها تصویر دارند، اما هویت ندارند.

خروجی AI اگر هدایت نشود، زیباست ولی بی‌برند و بی‌هدف است.

این فاز به تیم یاد می‌دهد:

چطور AI تصویرساز را از «نقاش ناشی» به «طراح برندمحور» تبدیل کند.


مسئله‌ای که این فاز حل می‌کند
مشکل رایج:

تصویر هست، ولی اگر لوگو را برداری، نمی‌فهمی مال چه برندی است

تصاویر قشنگ‌اند، اما پیام ندارند

خروجی‌ها شبیه هم‌اند (همان سبک‌های تکراری AI)

ریشه مشکل:

AI تصویرساز سلیقه ندارد
AI فقط دستور اجرا می‌کند، نه تصمیم برند.

پس:

اگر پرامپت ضعیف باشد → خروجی بی‌هویت است
اگر پرامپت دقیق باشد → خروجی برندمحور است





## تعریف Visual Prompting به زبان ساده

**Visual Prompting یعنی:**
نوشتن دستور دقیق برای اینکه AI:

* چه چیزی را بکشد
* با چه حس و سبکی
* در چه نوری
* با چه عناصر برند

در واقع:

> پرامپت تصویری = بریف طراحی حرفه‌ای

---

## اجزای اصلی پرامپت تصویری (خیلی مهم)

هر پرامپت تصویری حرفه‌ای باید این ۵ بخش را داشته باشد:

---

### ۱. Subject (سوژه)

**سوژه یعنی: دقیقاً چه چیزی در تصویر است؟**

❌ اشتباه:

> a restaurant poster

✅ درست:

> a modern fast-food restaurant counter with a smiling staff member holding a branded burger box

سوژه باید:

* مشخص
* قابل تصور
* بدون ابهام باشد

---

### ۲. Style (سبک)

**سبک یعنی تصویر شبیه چه چیزی باشد؟**

مثال‌ها:

* minimal
* cinematic
* luxury
* street photography
* illustration
* 3D render

سبک تعیین می‌کند:

* تصویر رسمی است یا جوان؟
* لوکس است یا دوستانه؟
* هنری است یا تجاری؟

---

### ۳. Mood (حس)

**Mood یعنی تصویر چه احساسی منتقل کند؟**

مثال‌ها:

* energetic
* warm
* premium
* fun
* calm
* bold

این بخش مستقیماً به **احساس برند** وصل است.

---

### ۴. Lighting (نور)

نور یعنی:

* تصویر روشن است یا تاریک؟
* دراماتیک است یا ساده؟
* طبیعی است یا استودیویی؟

مثال:

* soft natural light
* cinematic dramatic lighting
* bright daylight
* warm ambient light

نور اشتباه = حس اشتباه
نور درست = برند درست

---

### ۵. Brand Elements (عناصر برند)

مهم‌ترین بخش.

شامل:

* رنگ‌های برند
* فضا (مدرن، سنتی، صنعتی، لوکس…)
* vibe برند

مثال:

> using brand colors red and black, modern urban atmosphere, clean composition

اینجاست که تصویر «مال شما» می‌شود.

---

## چطور با پرامپت تصویری خروجی بسازند؟

### ۱. پوستر تبلیغاتی

هدف:

* انتقال پیام
* جلب توجه
* هویت برند

پرامپت باید:

* سوژه واضح
* حس قوی
* فضای تبلیغاتی داشته باشد

---

### ۲. بنر تبلیغاتی

هدف:

* سریع دیده شود
* پیام ساده
* قابل استفاده در سایت یا تبلیغات

پرامپت:

* ساده‌تر
* تمرکز روی یک عنصر
* فضای خالی (negative space)

---

### ۳. کاور شبکه اجتماعی

هدف:

* توقف اسکرول
* انتقال حس برند
* هماهنگی با بقیه پست‌ها

پرامپت:

* قوی در Mood
* سازگار با لحن برند
* قابل تکرار در سری محتوا

---

## تمرین عملی (بخش بسیار مهم)

### تمرین:

**یک برند → ۳ کانسپت تصویری کاملاً متفاوت**

مثلاً برای یک برند غذایی:

1. کانسپت احساسی (Warm & Friendly)
2. کانسپت جوان و پرانرژی (Bold & Modern)
3. کانسپت لوکس (Premium & Clean)

برای هر کانسپت:

* سوژه فرق دارد

* اما **برند ثابت می‌ماند**

این تمرین مغز تیم را از «تولید تصویر» به «طراحی کانسپت» می‌برد.

## خروجی این فاز چیست؟

بعد از این فاز، تیم می‌فهمد که:

1. AI تصویرساز **طراح نیست**

و نتیجه‌ی طلایی:

> **AI تصویرساز = طراح بی‌سلیقه‌ای که باید هدایت شود**

کسی که پرامپت بلد است:

* کنترل دارد

بعد از فاز Marketing Core، تیم شما حالا می‌تواند متن‌های هدفمند و با لحن برند بسازد.
اما تولید محتوا برای شبکه‌های اجتماعی نیاز به سفارشی‌سازی برای هر پلتفرم دارد.

شبکه‌های اجتماعی هر کدام رفتار کاربر متفاوت دارند:

لینکدین: حرفه‌ای، تحلیلی، آموزنده

## ۱. مسئله‌ای که حل می‌کنیم
شاید کپی شما خوب باشد، اما **با پلتفرم شبکه اجتماعی هم‌خوانی ندارد**.  
یعنی محتوا:  
- اسکرول کاربران را متوقف نمی‌کند  
- تعامل ایجاد نمی‌کند  
- لحن برند یا فرمت مناسب پلتفرم را ندارد  

هدف این فاز: تولید محتوا برای شبکه‌های اجتماعی که
**هم جذاب، هم تعامل‌ساز و هم برندمحور باشد**.

---

## ۲. آموزش پرامپت بر اساس پلتفرم

- **اینستاگرام** → احساسی، کوتاه، جذاب، دیداری  
- **لینکدین** → تحلیلی، حرفه‌ای، دقیق، ارزش‌آفرین  
- **واتس‌اپ بیزینس** → مستقیم، انسانی، شخصی‌سازی‌شده  

### انواع محتوا
- **کپشن** → متن کوتاه و اثرگذار  
- **ریلز** → Hook (جلب توجه) + Script (متن اصلی و داستان)  
- **استوری تعاملی** → نظرسنجی، سوال، بازی کوچک  
- **کامنت‌سازی هوشمند** → ایجاد تعامل و پاسخ‌دهی خودکار  


| پلتفرم         | لحن و سبک                   | نوع محتوا                                    |
| -------------- | --------------------------- | -------------------------------------------- |
| اینستاگرام     | احساسی، کوتاه، جذاب         | کپشن، ریلز، استوری تعاملی                    |
| لینکدین        | تحلیلی، حرفه‌ای، ارزش‌آفرین | کپشن حرفه‌ای، مقاله کوتاه، ریلز تحلیلی       |
| واتس‌اپ بیزینس | مستقیم، انسانی، شخصی        | پیام مستقیم، استیکر/پیام تعاملی، پاسخ هوشمند |


---

## ۳. نکته طلایی
یک پرامپت شبکه اجتماعی موفق باید:  
1. اسکرول کاربران را متوقف کند  
2. تعامل بسازد (لایک، کامنت، اشتراک‌گذاری)  
3. هویت و لحن برند را حفظ کند  

> این بخش عملی‌ترین قسمت دوره است؛ اینجاست که کپی‌های شما تبدیل به **کمپین واقعی و پول‌ساز** می‌شوند.

---

## ۴. تمرین عملی
**هدف:** طراحی یک **کمپین ۷ روزه فقط با پرامپت**  

مراحل تمرین:  
1. انتخاب یک برند از لیست ۸ برند  
2. مشخص کردن پلتفرم هدف (اینستاگرام، لینکدین، واتس‌اپ)  
3. طراحی محتوای روزانه شامل:  
   - کپشن  
   - ریلز (Hook + Script)  
   - استوری تعاملی  
   - کامنت‌سازی هوشمند  
4. پالایش و اطمینان از ثبات لحن برند و ایجاد تعامل  

> می‌توانید این تمرین را در **Google Sheet یا Excel** انجام دهید و هر روز یک ردیف به یک محتوا اختصاص دهید.

---

## ۵. خروجی فاز
- یک کمپین کامل و ۷ روزه بچیند  
- محتوای شبکه اجتماعی بسازد که
**نه تنها پست تکی بلکه جریان برند و تعامل واقعی ایجاد می‌کند**  
- لحن و هویت برند را در تمام پست‌ها حفظ کند
