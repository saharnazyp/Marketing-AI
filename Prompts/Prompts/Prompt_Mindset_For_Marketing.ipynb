Why Phase 0 Exists

Most prompt failures are not technical failures.
They are thinking failures.

This phase exists to:

Break the myth of AI as a creative brain

Replace “asking” with “briefing”

Establish responsibility for output quality

Skipping this phase guarantees shallow results in later phases.

03_command_vs_scenario.md
Command vs Scenario

Command (Wrong)

```python
Write an Instagram caption for a restaurant.
```

Scenario (Correct)

```python
You are a brand copywriter for a casual dining restaurant.
Target audience: 20–30 year olds.
Tone: friendly, energetic, not salesy.
Goal: increase engagement.
Output: max 2 lines.
```

04_ai_and_creativity.md
Creativity Is Not Outsourced

AI does not create ideas.
It expands existing ones.

Bad idea × AI = polished nonsense
Good idea × AI = scalable creativity

AI amplifies direction, not imagination.

05_prompting_as_thinking.md
Prompting Is Thinking Out Loud

A good prompt is evidence of:

clear intent

audience awareness

brand discipline

Prompting is a thinking skill disguised as a writing skill.

Treat each episode like a campaign

Ask:

What was the goal of the campaign?

Where was the KPI?

If we had AI, where would it help?

Where would it be a disaster in the real world?

Professional summary

Emily in Paris is like fast food:

Tasty

Visually inspiring

But if that’s all the team’s mental diet is, they’ll be malnourished

Next to it should be:

Mad Men → Depth

Moneyball → Data

Social dilemma ← Ethics

Emily in Paris → Today’s culture

With this combination, the team both understands the trend and doesn’t get fooled by it.
