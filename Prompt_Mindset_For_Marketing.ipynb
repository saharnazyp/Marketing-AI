Why Phase 0 Exists

Most prompt failures are not technical failures.
They are thinking failures.

This phase exists to:

Break the myth of AI as a creative brain

Replace “asking” with “briefing”

Establish responsibility for output quality

Skipping this phase guarantees shallow results in later phases.

03_command_vs_scenario.md
Command vs Scenario

Command (Wrong)

```python
Write an Instagram caption for a restaurant.
```

Scenario (Correct)

```python
You are a brand copywriter for a casual dining restaurant.
Target audience: 20–30 year olds.
Tone: friendly, energetic, not salesy.
Goal: increase engagement.
Output: max 2 lines.
```

Creativity Is Not Outsourced

AI does not create ideas.
It expands existing ones.

Bad idea × AI = polished nonsense
Good idea × AI = scalable creativity

AI amplifies direction, not imagination.

05_prompting_as_thinking.md
Prompting Is Thinking Out Loud

A good prompt is evidence of:

clear intent

audience awareness

brand discipline

Prompting is a thinking skill disguised as a writing skill.

Treat each episode like a campaign

Ask:

What was the goal of the campaign?

Where was the KPI?

If we had AI, where would it help?

Where would it be a disaster in the real world?

Professional summary

Emily in Paris is like fast food:

Tasty

Visually inspiring

But if that’s all the team’s mental diet is, they’ll be malnourished

Next to it should be:

Mad Men → Depth

Moneyball → Data

Social dilemma ← Ethics

Emily in Paris → Today’s culture

With this combination, the team both understands the trend and doesn’t get fooled by it.

prompt_components
Components of a Professional Prompt

Role / Persona
AI needs to understand who it is.
Example: Copywriter, social media manager, brand strategist.

Goal / Objective
What is the output meant to achieve?
Example: Increase engagement, generate leads, educate audience.

Context / Background
Brand identity, audience type, product details, platform.

Constraints / Limitations
Tone, style, length, forbidden words, brand rules.

Output Format
List, paragraph, caption, script, etc.

dos_and_donts.md
Dos

Be explicit with audience and purpose

Specify tone and style

Provide platform or format

Don’ts

One-line vague instructions

Assuming AI knows your brand

Skipping constraints

Example:

Bad Prompt:

```python
Write a marketing caption for our restaurant.
```

```python
You are a social media copywriter for a fast-casual urban restaurant.
Audience: 20–30 year olds
Tone: friendly, witty, professional
Goal: increase Instagram engagement
Length: max 2 lines
```

prompt_iterations.md
Iterative Prompting

Often, the first prompt does not give perfect output.
Iteration = key to professional prompting.

Steps

Draft initial prompt

Evaluate output

Refine prompt with additional context, examples, or constraints

Repeat until output aligns with goal

Tip:
Use multi-step prompting for complex tasks:

Step 1: Generate ideas

Step 2: Refine ideas

Step 3: Final output in desired format
