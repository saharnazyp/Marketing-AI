## ۲. پرامپت برای کپی فروش (Copywriting)

### ساختار محتوا
1. **تیتر (Headline)**: کوتاه، جذاب، مستقیم  
2. **بدنه (Body)**: توضیح مزیت یا محصول، نشان دادن نیاز مخاطب  
3. **CTA (Call To Action)**: مستقیم و واضح، مخاطب را به اقدام دعوت می‌کند
